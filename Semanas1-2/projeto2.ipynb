```python
import spacy
import nltk
from nltk.corpus import stopwords
from nltk.stem import RSLPStemmer
from nltk.probability import FreqDist
import unicodedata
import string

nltk.download('stopwords')
nltk.download('rslp')

nlp = spacy.load("pt_core_news_lg")
```

Ler arquivo txt com o texto

```python
caminho_arquivo = "/content/texto.txt"

with open(caminho_arquivo, "r", encoding="utf-8") as f:
    texto_original = f.read()

print("Texto original:\n")
print(texto_original)
print("\n\nTamanho do texto (em palavras):", len(texto_original.split()))
```

Classifica√ß√£o Gramatical

```python
doc = nlp(texto_original)

print("--- Classes Gramaticais (POS) e Lematiza√ß√£o ---")
print(f"{'Token'.ljust(15)} | {'Classe (POS)'.ljust(15)} | {'Lema'.ljust(15)}")
print("-" * 47)

for token in doc:
    print(f"{token.text.ljust(15)} | {token.pos_.ljust(15)} | {token.lemma_.ljust(15)}")
```

Remo√ß√£o de Stopwords, Pontua√ß√£o e N√∫meros


```python
stop_words_nltk = stopwords.words('portuguese')

lemmas_filtrados = []

for token in doc:
    if (not token.is_stop and
        not token.is_punct and
        not token.is_digit and
        not token.is_space):

        lemmas_filtrados.append(token.lemma_)

print("--- Texto Ap√≥s Remo√ß√£o de Stopwords, Pontua√ß√£o, N√∫meros e Espa√ßos ---")
print(lemmas_filtrados)
```

Normaliza√ß√£o

```python
def normalizar(texto):
    texto = texto.lower()

    texto_sem_acento = unicodedata.normalize('NFKD', texto)\
                                  .encode('ASCII', 'ignore')\
                                  .decode('utf-8')
    return texto_sem_acento

tokens_finais_normalizados = [normalizar(lemma) for lemma in lemmas_filtrados]

print("--- Tokens Finais (Lematizados e Normalizados) ---")
print(tokens_finais_normalizados)
```

Contagem de palavras frequentes

```python
freq = FreqDist(tokens_finais_normalizados)

print("--- 10 Palavras Mais Frequentes ---")
print(freq.most_common(10))
```

Stemming

```python
stemmer = RSLPStemmer()

tokens_para_stemming = []
for token in doc:
    if (not token.is_stop and
        not token.is_punct and
        not token.is_digit):

        token_norm = normalizar(token.text)
        token_stem = stemmer.stem(token_norm)
        tokens_para_stemming.append(token_stem)

print("--- Tokens Finais (Stemming + Normalizados) ---")
print(tokens_para_stemming)
```

An√°lise de Sentimentos

```python
import pandas as pd
import numpy as np

lexico = {
    # --- Originais ---
    "bom": 1, "√≥timo": 2, "excelente": 3, "divertido": 1, "maravilhoso": 3,
    "ruim": -1, "p√©ssimo": -3, "terr√≠vel": -3, "horr√≠vel": -2,
    "gostar": 1, "odiar": -3,
    "tranquilo": 1,
    "positivo": 1,
    "calmo": 1,
    "feliz": 2,
    "agrad√°vel": 1,
    "gratificante": 2,
    "animado": 1,
    "interessante": 1,
    "saboroso": 1,
    "sorrir": 1,
    "perfeito": 3,
    "apoio": 1,
    "tenso": -1,
    "estresse": -2,
    "confuso": -1,
    "mal": -1,
    "r√≠spido": -2,
    "triste": -2,
    "dificuldade": -1,
    "cancelado": -1
}
emojis = {"üòä": 2, "üòÅ": 2, "üòÇ": 1, "üò¢": -2, "üò°": -3, "üò†": -2}
intensificadores = {
    "muito": 1.5, "extremamente": 1.5, "super": 1.5, "bastante": 1.5,
    "t√£o": 1.5, "mais": 1.2
}
atenuadores = {
    "pouco": 0.5, "quase": 0.5, "meio": 0.5, "menos": 0.8
}
negacoes = {"n√£o", "nunca", "jamais", "nenhum", "nem"}

def calcular_sentimento_final(tokens):
    dados_analise = []
    score_total_simples = 0

    for i, token in enumerate(tokens):
        score_base = 0
        tipo = "neutro"

        if token in lexico:
            score_base = lexico[token]
            tipo = "palavra"
        elif token in emojis:
            score_base = emojis[token]
            tipo = "emoji"

        if score_base == 0:
            dados_analise.append({"token": token, "score": 0, "tipo": "neutro", "modificador": None})
            continue

        score_final = score_base
        modificador = None

        window = tokens[max(0, i-3):i]
        if any(w in negacoes for w in window):
            score_final *= -1
            modificador = "negacao"

        if i > 0:
            last_word = tokens[i-1]
            if last_word in intensificadores:
                score_final *= intensificadores[last_word]
                modificador = "intensif" if not modificador else "negacao+int"
            elif last_word in atenuadores:
                score_final *= atenuadores[last_word]
                modificador = "atenuador" if not modificador else "negacao+atn"

        dados_analise.append({
            "token": token,
            "score": score_final,
            "tipo": tipo,
            "modificador": modificador
        })
        score_total_simples += score_final

    df = pd.DataFrame(dados_analise)

    df_ativos = df[df['score'] != 0].copy()

    soma_simples = score_total_simples

    if not df_ativos.empty:
        media_intensidade = df_ativos['score'].mean()
        pesos = np.arange(1, len(df_ativos) + 1)
        soma_ponderada = np.sum(df_ativos['score'].values * pesos)
        soma_ponderada_normalizada = soma_ponderada / (len(df_ativos) / 2)
    else:
        media_intensidade = 0
        soma_ponderada_normalizada = 0

    resultado = {
        "dataframe": df,
        "dataframe_ativos": df_ativos,

        "metricas": {
            "Soma_simples": soma_simples,
            "Media_intensidade": media_intensidade,
            "Soma_ponderada": soma_ponderada_normalizada,

            "qtd_tokens_total": len(df),
            "qtd_tokens_sentimento": len(df_ativos)
        }
    }

    return resultado
```
